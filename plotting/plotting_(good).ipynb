# 2 Comparing the two ‘dialects’

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### pyplot Interface

```python
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()
```

### OO Interface

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend()
#In summary, the first method directly uses the 
#plt.plot() and plt.legend() functions, while the 
#second method involves creating a figure and axes objects explicitly, 
#providing more control over the plot's properties and layout.
```

```python
x = np.linspace(0, 2*np.pi, 100)
sin_x = np.sin(x)
cos_x = np.cos(x)

# Create a figure and axes object
fig, ax = plt.subplots()

# Plot sin(x) and cos(x) on the same axes
ax.plot(x, sin_x, label='sin x')
ax.plot(x, cos_x, label='cos x')

# Add legend to the plot
ax.legend()

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y', rotation=0)
ax.set_title('Plot of sin(x) and cos(x)')
plt.show()
```

# 3 What is OO ax, really?

```python
fig, ax = plt.subplots(nrows=2, ncols=1) #two rows and one column 

ax[0].plot(x, cos_x, label='cos x') #ax is a Numpy array with 2 'things' containing matplotlib axis
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
```

# 4 A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()
```

```python
tan_x = np.tan(x)

fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, tan_x, label=r'$\tan(x)$')
ax[1].fill_between(x, 0, tan_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\tan(x)$ and $\cos(x)$')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5), #specify size of plots
                       sharex=True) #share x axis


ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25) #fill the plots
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax: #for loop to iterates through the items in the ax 
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)


ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')

fig.tight_layout()
```

# 5 Multiple rows and columns

```python
# Create a figure with 2x2 subplots arranged in a grid
# sharex='col' and sharey='col' ensures that the x-axis and y-axis are shared across columns
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(5, 5), sharex='col', sharey='col')

# Access the individual subplots for better readability
top_left, top_right, bottom_left, bottom_right = ax.flatten()

# Set titles for each subplot
top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

# Add horizontal lines to each subplot
top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

# Add grid lines to each subplot with alpha=0.25
for a in ax.flatten():
    a.grid(alpha=0.25)

# Adjust layout to prevent overlap of subplots
plt.tight_layout()
```

```python
# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with 2x2 subplots arranged in a grid
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8), sharex='col', sharey='row')

# Access the individual subplots for better readability
top_left, top_right, bottom_left, bottom_right = ax.flatten()

# Plot sin(x) on the top-left subplot
top_left.plot(x, y1, color='blue')
top_left.set_title('Sin(x)')
top_left.set_ylim(-1.5, 1.5)  # Adjust y-axis limits for sin(x)

# Plot cos(x) on the top-right subplot
top_right.plot(x, y2, color='red')
top_right.set_title('Cos(x)')
top_right.set_ylim(-1.5, 1.5)  # Adjust y-axis limits for cos(x)

# Plot tan(x) on the bottom-left subplot
bottom_left.plot(x, y3, color='green')
bottom_left.set_title('Tan(x)')
bottom_left.set_ylim(-10, 10)  # Adjust y-axis limits for tan(x)

# Add grid lines to each subplot with alpha=0.25
for a in ax.flatten():
    a.grid(alpha=0.25)

# Adjust layout to prevent overlap of subplots
plt.tight_layout()

plt.show()

```

### A quick helper function

```python
def get_plot_data():
    # Initialize an empty dictionary to store the data
    data = {}
    
    # File path for the data containing all graduates
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    # Load the data from the CSV file, skipping the first row (assuming it contains headers)
    # Store the data in the dictionary with key 'All'
    data['All'] = np.loadtxt(filename, skiprows=1)

    # File path for the data containing NUS graduates
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    # Load the data from the CSV file, skipping the first row
    # Store the data in the dictionary with key 'NUS'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    # Return the dictionary containing the loaded data
    return data
```

### The histogram

```python
# Set the style of the plot using the 'bmh' style
plt.style.use('bmh')

# Call the get_plot_data() function to retrieve the data
data = get_plot_data()

# Create a histogram plot using plt.hist() for the data of all graduates and NUS graduates
# bins=50 specifies to split the data into 50 bins
# label=['All', 'NUS'] provides labels for the legend
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])

# Set the x-axis label
plt.xlabel('Mean of Basic Monthly Earning (S$)')

# Set the y-axis label
plt.ylabel('Number of Students')

# Add a legend to the plot
plt.legend()
```

```python
# Generate some sample data
data1 = np.random.normal(loc=0, scale=1, size=1000)  # Data for group 1
data2 = np.random.normal(loc=2, scale=1, size=1000)  # Data for group 2

# Create a histogram plot with custom styling
plt.style.use('bmh')  # Set the style of the plot

# Plot the histogram for group 1 with blue bars
plt.hist(data1, bins=30, color='blue', alpha=0.5, label='Group 1')

# Plot the histogram for group 2 with orange bars
plt.hist(data2, bins=30, color='orange', alpha=0.5, label='Group 2')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Two Groups')
plt.legend()

plt.show()
```

## 6.2 Scatter plots

```python
# Set the style of the plot to "seaborn-v0_8-darkgrid"
plt.style.use("seaborn-v0_8-darkgrid")

# Call the get_plot_data() function to retrieve the data
data = get_plot_data()

# Iterate over the data dictionary, which contains labels as keys and corresponding numbers as values
for label, numbers in data.items():
    # Generate x-values as a range of indices for the data points
    x = range(len(numbers))
    # Extract y-values from the data
    y = numbers
    # Create a scatter plot for the current label and data
    plt.scatter(x, y, label=label, alpha=.5)

# Set the x-axis label
plt.xlabel('Position in the list')
# Set the y-axis label
plt.ylabel('Mean of Basic Monthly Earning (S$)')
# Add a legend to the plot
plt.legend()
```

```python
# Generate sample data for two groups
x_values = np.linspace(0, 10, 100)  # x-values are evenly spaced between 0 and 10
y_values_group1 = np.sin(x_values) + np.random.normal(loc=0, scale=0.1, size=100)  # y-values for group 1
y_values_group2 = np.cos(x_values) + np.random.normal(loc=0, scale=0.1, size=100)  # y-values for group 2

# Set the style of the plot
plt.style.use("seaborn-v0_8-darkgrid")

# Plot the data for group 1
plt.scatter(x_values, y_values_group1, label='Group 1', color='blue', alpha=0.7)

# Plot the data for group 2
plt.scatter(x_values, y_values_group2, label='Group 2', color='red', alpha=0.7)

# Add labels and legend
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot of Two Groups')
plt.legend()

plt.show()
```

## 6.3 Bar charts

```python
# Define a dictionary containing student numbers for each major
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}

# Extract the keys (majors) and values (numbers) from the dictionary
majors = list(student_numbers.keys())  # List of majors
numbers = list(student_numbers.values())  # List of student numbers corresponding to each major

# Set the style of the plot to 'ggplot'
plt.style.use('ggplot')

# Create a bar plot using plt.bar()
# The x-values are the majors, and the y-values are the corresponding student numbers
plt.barh(majors, numbers)

# Add labels for the x-axis and y-axis
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

```python
import matplotlib.pyplot as plt

# Define data for different products and their respective sales numbers
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [120, 90, 150, 110]  # Sales numbers for each product

# Set the style of the plot to 'ggplot'
plt.style.use('ggplot')

# Create a bar plot for the sales data
plt.bar(products, sales, color='skyblue')

# Add labels for the x-axis and y-axis
plt.xlabel('Products')
plt.ylabel('Sales Numbers')

# Add a title to the plot
plt.title('Sales Numbers for Different Products')

# Show the plot
plt.show()
```

## 6.4 Pie charts

```python
# Define a dictionary containing student numbers for each major
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}

# Extract the keys (majors) and values (numbers) from the dictionary
majors = list(student_numbers.keys())  # List of majors
numbers = list(student_numbers.values())  # List of student numbers corresponding to each major

# Set the style of the plot to 'fivethirtyeight'
plt.style.use('fivethirtyeight')

# Create a pie chart using plt.pie()
# The data is represented by 'numbers' and the labels are 'majors'
# autopct='%1.1f%%' formats the percentages to one decimal place
# startangle=-90 rotates the pie chart by 90 degrees counter-clockwise to start from the left
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=-90                
        )

# Add a title to the pie chart
plt.title('Percentage of each major')
```

```python
import matplotlib.pyplot as plt

# Define data for different categories and their respective values
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [30, 20, 25, 15]  # Values for each category

# Set the style of the plot to 'fivethirtyeight'
plt.style.use('fivethirtyeight')

# Create a pie chart for the data
# The data is represented by 'values' and the labels are 'categories'
# autopct='%1.1f%%' formats the percentages to one decimal place
# startangle=90 rotates the pie chart by 90 degrees clockwise to start from the top
plt.pie(values, 
        labels=categories,
        autopct='%1.1f%%',   # Format the percentages
        startangle=90        # Start the first slice at 90 degrees (top of the circle)
        )

# Add a title to the pie chart
plt.title('Distribution of Categories')

# Display the plot
plt.show()
```
