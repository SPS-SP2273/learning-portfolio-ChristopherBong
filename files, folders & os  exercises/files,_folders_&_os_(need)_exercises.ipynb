```python
def rename_my_file(old_file_name):
    month_dictionary = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
} #dictionary containing months and their numerical values

    parts = original_filename.split('-') #split the filename after the '-'

    month = parts[0]
    date, city = parts[1].split('_')

    month_numeric = month_dictionary[month] #covert month to numerical values 

    new_filename = f"{month_numeric}-{date}_{city}" 

    return new_filename
```

```python
list_files = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
             'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
             'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
             'jan-01_tokyo.txt'] #list of filenames

def rename_my_file(old_file_name):
    month_dictionary = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
} #dictionary containing months and their numerical values 

    parts = old_file_name.split('-') #split the filename after the '-'

    month = parts[0]
    date, city = parts[1].split('_') #split the filename after the '_'

    month_numeric = month_dictionary[month] #covert month to numerical values 

    new_filename = f"{month_numeric}-{date}_{city}" 

    return new_filename

for old_file_name in list_files:
    new_file_name = rename_my_file(old_file_name)
    print(f"{old_file_name} ----> {new_file_name}")
```

```python
directory = r'C:\Users\chris\OneDrive\Documents\GitHub\learning-portfolio-ChristopherBong\files, folders & os  exercises\os-collaboration-exercise-data'
#directory containing files

txt_files = glob.glob(os.path.join(directory, '*.txt')) #get all the .txt files in the directory 

for file_path in txt_files: #rename all .txt files
    file_name = os.path.basename(file_path)
    if file_name != rename_my_file(file_name):
        new_file_name = rename_my_file(file_name) #file hasn't been renamed
        os.rename(os.path.join(directory, file_name), os.path.join(directory, new_file_name)) #rename
        print(f"Renamed: {file_name} ----> {new_file_name}")
    else:
        print(f"Skipping: {file_name} (Already renamed)") #already renamed
```

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

main_directory = r'C:\Users\chris\OneDrive\Documents\GitHub\learning-portfolio-ChristopherBong\files, folders & os  exercises\os-collaboration-exercise-data'
#location of where the new folders will be made 

for city in cities: #for loop to create folders for each city
    city_directory = os.path.join(main_directory, city)
    if not os.path.exists(city_directory): #check if folder exist 
        os.makedirs(city_directory) #create folder
        print(f"Created folder for {city}")
    else:
        print(f"Folder for {city} already exists")
```

```python
cairo_files = glob.glob('os-collaboration-exercise-data\*cairo*') #list files with cairo

cairo_directory = 'os-collaboration-exercise-data\Cairo'
#directory to copy the cairo files to 

for file_path in cairo_files: # copy all the Cairo files to the Cairo directory
    file_name = os.path.basename(file_path)
    shutil.copy(file_path, cairo_directory)
    print(f"Copied {file_name} to {cairo_directory}")
```

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']


main_directory = r'C:\Users\chris\OneDrive\Documents\GitHub\learning-portfolio-ChristopherBong\files, folders & os  exercises\os-collaboration-exercise-data'

for city in cities:
    city_files = glob.glob(os.path.join(main_directory, f'*{city.lower()}*'))  # List all the files related to the current city
    

    city_directory = os.path.join(main_directory, city) # Directory where the city files will be moved


    for file_path in city_files: #move files to city folders
        file_name = os.path.basename(file_path)
        destination_file_path = os.path.join(city_directory, file_name)
        if not os.path.exists(destination_file_path): #Check if the file already exists 
            shutil.move(file_path, city_directory)
            print(f"Moved {file_name} to {city_directory}")
        else:
            print(f"File {file_name} already exists in {city_directory}. Skipping...")
```
